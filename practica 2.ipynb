Plotting sin linea:Para trazar solo los marcadores, pueden usar notaciones de cadena de acceso directo'o', que significa "anillos".Ejemplo: Dibuja dos puntos en el diagrama, uno en la pocision (1,3)y el otro en la pocision (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

puntosX = np.array([1,8])
puntosY = np.array([3,10])

plt.plot(puntosX,puntosY,'o')
plt.show()
```

Diagramas de dispersion (scretter)
Se emplea para mostrar la relacion entre 2  numericas, detectando correlaciones o agrupamientos,
Ejemplo:

```python
import matplotlib.pyplot as plt

altura = [150,160,170,180,190,200]
peso = [50,60,65,80,85,95]

plt.scatter(altura,peso,color = "red",marker = "X")
plt.title("Relacion entre la altura y el peso")
plt.xlabel("Altura(cms)")
plt.ylabel("Peso(kg)")
plt.show()
```

Graficos de barras(bar y barh)
Sirven para comparar categorias o variables tanto en orirntacion vertical como orizontal.
Ejemplo de barras verticales:

```python
import matplotlib.pyplot as plt

categotias = ["A","B","C","D"]
valores = [4,7,1,8]

plt.bar(categotias,valores, color ="orange")
plt.title("Grafico de barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
```
